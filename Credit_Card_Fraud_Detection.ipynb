**Undersampling**

**Building a sample dataset containing similar distribution of normal transaction and fraud transaction**

Number of fraudulent transaction is --> 492

```python
legit_sample = legit.sample(n = 492)
```

# Concatenating 2 DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.groupby('Class').mean()
```

#Splitting the data into Train and Test Data

```python
X = new_dataset.drop('Class', axis = 1)
Y = new_dataset['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 2)
```

```python
print(X)
```

```python
print(Y)
```

**Splitting the Data**

```python
print(X_train.shape, X_test.shape, X.shape)
```

**Model Training**

#**Logistic Regression**

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

**Model Evaluation**

**Accuracy Score**

```python
# Accuracy Score for Train data
X_train_predict = model.predict(X_train)
train_data_accuracy = accuracy_score(X_train_predict, Y_train)
print("The Accuracy Score for the train data is:",train_data_accuracy)
```

```python
# Accuracy Score for Test data
X_test_predict = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predict, Y_test)
print("The Accuracy Score for the test data is:",test_data_accuracy)
```
